```python
import matplotlib.pyplot as plt

marks = [3,8,19,12,14,21,1,16,11,6,22,11,28,17,28,15,30,20,10,8]
grade_intervals=[0,5,10,15,20,25,30]

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'red','size':15}

plt.title("Histogram of Marks",fontdict=font1)
plt.xlabel("Marks obtained",fontdict=font2)
plt.ylabel("No.of students",fontdict=font2)

plt.hist(marks,grade_intervals, color='orange')
plt.show
```
