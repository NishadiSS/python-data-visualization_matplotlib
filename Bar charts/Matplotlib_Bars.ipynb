## Draw 4 bars:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,color="red",width = 0.2)
plt.show()
```

## Draw 4 horizontal bars:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y,height = 0.1)
plt.show()
```

# Broken Barh

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()

font = {'family':'serif','color':'red','size':12}

ax.broken_barh([(2, 30), (70, 10)], (10, 9), facecolors='tab:purple')
ax.broken_barh([(10, 50), (100, 20), (130, 10)], (25, 9), facecolors=('tab:orange', 'tab:green', 'tab:blue'))
ax.set_ylim(0, 40)
ax.set_xlim(0, 200)
ax.set_xlabel('minutes from start', fontdict = font)
ax.set_yticks([15, 30], labels=['Group 1', 'Group 2'], fontdict = font)    
ax.grid(True)                                       


plt.show()
```
