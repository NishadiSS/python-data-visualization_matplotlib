```python
import matplotlib.pyplot as plt

colors=['Orange','Blue','Pink','Yellow','Red']
NoOfStudent=[75,100,25,50,40]

width=[0.2,0.5,0.3,0.6,0.5]

font1={'family':'serif','color':'green','size':18}
font2={'family':'serif','color':'darkred','size':15}

plt.bar(colors,NoOfStudent,width=width,color=colors)

plt.xlabel('Colors',fontdict=font2)
plt.ylabel('Number Of Student',fontdict=font2)
plt.title('Favourite Color Of Student',fontdict=font1)

plt.show()
```

```python
import matplotlib.pyplot as plt

colors=['Green','Darkred','Orange','Yellow','Purple']
NoOfStudent=[75,100,25,50,40]

height=[0.2,0.5,0.3,0.6,0.5]

font1={'family':'serif','color':'green','size':18}
font2={'family':'serif','color':'darkred','size':15}

plt.barh(colors,NoOfStudent,height=height,color=colors)

plt.ylabel('Colors',fontdict=font2)
plt.xlabel('Number Of Student',fontdict=font2)
plt.title('Favourite Color Of Student',fontdict=font1)

plt.show()

```
