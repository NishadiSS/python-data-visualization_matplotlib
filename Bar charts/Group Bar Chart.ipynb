```python
import matplotlib.pyplot as plt
import numpy as np

animals = ("Giraffes", "Monkeys","Orangutans")
places = {
    'SF zoo':(12,22,18),
    'LA zoo':(12,15,25),
    'DC zoo':(10,20,6),
}

x= np.arange(len(animals))
width=0.25 #the width of the bars
multiplier=0

font1 = {'family':'serif','color':'red','size':18}

fig,ax = plt.subplots(layout='constrained')

for attribute , count in places.items():
    offset = width * multiplier
    rects = ax.bar(x+offset,count,width,label=attribute)
    ax.bar_label(rects,padding=3)
    multiplier += 1
    
ax.set_ylabel('Count',fontdict=font1)
ax.set_xlabel('Animals',fontdict=font1)
ax.set_xticks(x+width,animals)

ax.legend(loc='upper left',ncols=3)
ax.set_ylim(0,30)

plt.show()
```
