```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

activities = ['Reading', 'Playing', 'Baking', 'Washing hands']
counts = [40, 100, 30, 55]
bar_labels = ['green', 'pink', '_green', 'red']
bar_colors = ['tab:green', 'tab:pink', 'tab:green', 'tab:red']

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

ax.bar(activities, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Number of children', fontdict = font2)
ax.set_xlabel('Type of activity', fontdict = font2)
ax.set_title('Activities at home', fontdict = font1)

ax.legend(title='Activity color')

plt.show()
```
