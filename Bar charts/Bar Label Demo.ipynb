```python
import matplotlib.pyplot as plt
import numpy as np

sports=('Baseball','Basketball','Football')
sex_counts = {
    'Male': np.array([54,34,61]),
    'Female': np.array([54,78,58]),
}
width = 0.4 #the width of the bars

fig, ax = plt.subplots()
bottom=np.zeros(3)

font1={'family':'serif','color':'blue','size':20}
colors={'Male':'green','Female':'yellow'}

for sex, sex_count in sex_counts.items():
    p=ax.bar(sports,sex_count,width,label=sex,bottom=bottom,color=colors[sex])
    bottom+=sex_count
    
    ax.bar_label(p, label_type='center')
    
ax.set_title('What is Your Favorite Sports?',fontdict=font1)
ax.legend()

plt.show()
```
