```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[400,200,400,300,200,100,600]
plt.scatter(x,y,c=colors,s=sizes,marker ="*",edgecolor="y",linewidth=2)
plt.title("Scatter Plot",fontsize=15,color='red')
plt.xlabel("x-axis",fontsize=15,color='blue')
plt.ylabel("y-axis",fontsize=15,color='blue')
plt.show()
```
