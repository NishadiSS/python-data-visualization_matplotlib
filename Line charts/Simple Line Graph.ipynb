```python
import matplotlib.pyplot as plt
import numpy as np

Weeks = np.array([1,2,3,4,5,6,7,8])
No_Of_Books = np.array([60,120,120,90,90,60,90,10])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'green','size':15}

plt.plot(Weeks,No_Of_Books, linestyle='dotted',color = 'hotpink',linewidth=2,marker='o',markersize=8)

plt.title("Books Sold" , fontdict=font1,loc ='center')
plt.xlabel("Weeks",fontdict=font2)
plt.ylabel("No.of books sold" , fontdict=font2)

plt.show()
```
