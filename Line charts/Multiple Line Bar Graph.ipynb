```python
import matplotlib.pyplot as plt
Months=['Jan','Feb','Mar','Apr','May','Jun']
Tablets=[10,20,25,40,70,100]
Laptop = [33,28,38,58,52,60]
Desktop = [80,42,30,20,10,5]

font={'family':'serif','color':'red','size':20}

plt.plot(Months,Tablets,label='Tablets')
plt.plot(Months,Laptop,label='Laptop')
plt.plot(Months,Desktop,label='Desktop')

plt.title('Product Trends by Month',fontdict=font)
plt.legend()
plt.grid()
plt.show()
```
